## Introducción a análisis frecuencial con python

Las comunmente llamadas series de Fourier hace referencia a la representación de una función en forma de funciones trigonométricas. Una forma de esta representación fue propuesta por Fourier de la siguiente manera: $$f(x)=\frac{a_0}{2}+\sum_{n=1}^{\infty}{a_n\cos nx + b_n\sin nx}\quad x\in(-\pi,\pi]\qquad(1)$$
Dónde: $$a_n=\frac{1}{\pi}\int_{-\pi}^{\pi}{f(x)\cos nx dx}\qquad(2)$$
$$b_n=\frac{1}{\pi}\int_{-\pi}^{\pi}{f(x)\sin nx dx}\qquad(3)$$
#### La transformada de Fourier:
Hace referencia a una integral de transformada definida sobre la línea $x\in(-\infty,\infty)$ definida como:$$F(k)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}{e^{-ikx}f(x)dx}\qquad(4)$$
Y su transformada inversa:$$f(x)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}{e^{ikx}F(k)dk}\qquad(5)$$
Dónde $e^{\pm ikx}$ es el kernel de la transformada, el cual a su vez es el que describe el comportamiento oscilatorio de la función.
También, es posible observar que la función $F(k)$ es derivable en el dominio en donde la función $f(x)$ sea derivable:
$$F^{(n)}=(ik)^nF\qquad(6)$$

A continuación se puede observar un ejemplo de realizar la transformada de Fourier a una función dada $f(x)=e^{-x^2}$:

```python
L=20
n=128
fm=linspace(-n/2,n/2,n)
x2=linspace(-L/2,L/2,n+1)
x=x2[:-1]
u=exp(-(x**2))
ut=fft.fft(u)
utshift=fft.fftshift(ut)

figure(1)
plot(x,u)
xlabel('Time');ylabel('f(x)')
figure(2)
plot(fm,ut)
xlabel('Modos de Fourier');ylabel('F(k)')
figure(3)
plot(fm,utshift)
xlabel('Modos de Fourier');ylabel('F(k)')
title('Transformada de Fourier en su estado matemáticamente correcto')
```

En este caso se puede observar que el comando numpy.fft.fft proporciona la transformada rápida de Fourier de la función dada, sin embargo, es necesario utilizar el comando numpy.fft.fftshift para cambiar el componente de cero frecuencia al centro del espectro.

### Cálculo de Derivadas:
A continuación se presenta un cálculo de derivadas mediante el uso de la transformada de fourier y su comparación mediante el método de diferencias finitas de orden 2 y de orden 4.
Para recordar, el método de diferencias finitas está dado como:
#### Segundo Orden:
$$f'(x)=\frac{f(x+\Delta x)-f(x-\Delta x)}{2\Delta x}$$
#### Cuarto Orden:
$$f'(x)=\frac{-f(x+2h)+8f(x+h)-8f(x-h)+f(x-2h)}{12h}$$

En este caso, se presenta una comparación entre el cálculo de derivadas de una función sech($x$).

```python

dx=x[1]-x[0]
u=1/cosh(x)
ut=fft.fft(u)
a=linspace(0,n/2-1,n/2)
b=linspace(-n/2,-1,n/2)

#FFT derivatives
k=(2*pi/L)*concatenate((a,b),axis=0)
ut1=k*ut*1j
ut2=-k*k*ut
u1=fft.ifft(ut1).real; u2=fft.ifft(ut2).real 
u1Exact=-tanh(x)/cosh(x); u2Exact= (1/cosh(x))-(2/(cosh(x))**3)

#Finite difference

#2nd order accuracy
ux=linspace(0,n,n); ux[0]=(-3*u[0]+4*u[1]-u[2])/(2*dx)
for i in range (1,n-2):
    ux[i]=(u[i+1]-u[i-1])/(2*dx)
ux[n-1]=(3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx)

#4th order accuracy
ux2=linspace(0,n,n)
ux2[0]=(-3*u[0]+4*u[1]-u[2])/(2*dx); ux2[1]=(-3*u[1]+4*u[2]-u[3])/(2*dx)
for i in range (2,n-3):
    ux2[i]=(-u[i+2]+8*u[i+1]-8*u[i-1]+u[i-2])/(12*dx)
ux2[n-2]=(3*u[n-2]-4*u[n-3]+u[n-4])/(2*dx)
ux2[n-1]=(3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx)

#Comparisson:

figure(1)
plot(x,u,'r', label='$f(x)=sech(x)$')

plot(x,u1,'g', label='$f´_{fft}(x)$')
plot(x,u1Exact,'go',label='f´(x)=-tanh(x)sech(x)')

plot(x,u2,'b', label='$f´´_{fft}(x)$')
plot(x,u2Exact,'bo',label='f´´(x)=sech(x)-$2sech^3(x)$')
title('comparacion de calculo de derivadas mediante fft')
legend()

figure(2)
plot (x, u1Exact,'k',label='f´(x)'); plot(x,u1,'b^',label='Solución Fourier');
plot(x,ux,'r*',label='Diff. finitas 2° orden'); plot(x,ux2,'g*',label='Diff. Finitas 4° orden')
gca().set_xlim([-1.15,-0.75])
gca().set_ylim([0.47,0.5])
legend()

figure(3)
plot (x, u2Exact,'k',label='f´´(x)'); plot(x,u2,'b^',label='Solución Fourier');
#plot(x,ux,'r*',label='Diff. finitas 2° orden'); plot(x,ux2,'g*',label='Diff. Finitas 4° orden')
gca().set_xlim([-0.5,0.5])
gca().set_ylim([-1,-0.5])
legend()
```

Es posible observar que el método de cálculo de derivadas por transformada de Fourier es bastante bueno para este tipo de funciones, sin embargo, se debe tener en cuenta que esta función es periódica y contínua. Esto es muy importante debido a que por su caracter armónico, la transformada de Fourier no es la mejor opción a la hora de tratar con funciones ya sean no periódicas o discontínuas. Un ejemplo claro de esto se presenta a continuación evaluando la función tanh($x$):

```python
#tanh case:
u=tanh(x)
ut=fft.fft(u)
a=linspace(0,n/2-1,n/2)
b=linspace(-n/2,-1,n/2)

#FFT derivatives
k=(2*pi/L)*concatenate((a,b),axis=0)
ut1=k*ut*1j
ut2=-k*k*ut
u1=fft.ifft(ut1).real; u2=fft.ifft(ut2).real 
u1Exact=1/(cosh(x))**2; u2Exact= -2*tanh(x)/(cosh(x))**2

#Finite difference

#2nd order accuracy
ux=linspace(0,n,n); ux[0]=(-3*u[0]+4*u[1]-u[2])/(2*dx)
for i in range (1,n-2):
    ux[i]=(u[i+1]-u[i-1])/(2*dx)
ux[n-1]=(3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx)

#4th order accuracy
ux2=linspace(0,n,n)
ux2[0]=(-3*u[0]+4*u[1]-u[2])/(2*dx); ux2[1]=(-3*u[1]+4*u[2]-u[3])/(2*dx)
for i in range (2,n-3):
    ux2[i]=(-u[i+2]+8*u[i+1]-8*u[i-1]+u[i-2])/(12*dx)
ux2[n-2]=(3*u[n-2]-4*u[n-3]+u[n-4])/(2*dx)
ux2[n-1]=(3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx)

#Comparisson:
figure()

figure(4)
plot(x,u,'r', label='$f(x)=tanh(x)$')

plot(x,u1,'g', label='$f´_{fft}(x)$')
plot(x,u1Exact,'go',label='$f´(x)=sech^2(x)$')
plot(x,ux,'b^', label='$Diff. finitas 2° orden$')
plot(x,ux2,'k*', label='$Diff. finitas 4° orden$')
gca().set_ylim([-1.5,1.5])
title('comparacion de calculo de derivadas mediante fft')
legend()
```

En este caso es posible observar que dada la naturaleza no contínua, su espectro frecuencial es oscilante y asímismo al calcular la derivada de la función mediante este método, esta se comporta de manera oscilante. Mientras que al realizar el cálculo mediante diferencias finitas, se puede ver que el comportamiento es un poco más exacto.

### Filtro de señales
Otra aplicación que tiene la transformada de Fourier es para el filtrado de señales. Esta aplicación es bastante utilizada en la ingeniería puesto que al trazar una función en su dominio frecuencial, es posible eliminar mediante operaciones matemáticas simples las frecuencias parásitas presentes en una señal. A continuación se presentan un par de ejemplos donde se explica esto:

```python
#Filtering example

L=30; n=512
a=linspace(1,n/2,n/2-1)
b=linspace(-n/2,-1,n/2)
#FFT derivatives
k=(2*pi/L)*concatenate((a,b),axis=0)
t2=linspace(-L,L,n); t=t2[:-1]
u=1/cosh(t)
noise=5;ut=fft.fft(u)
utn=ut+noise*(numpy.random.rand(n-1)+1j*numpy.random.rand(n-1))
un=fft.ifft(utn)

figure(5)
plot(t,abs(un));xlabel('Tiempo');ylabel('|u|')
gca().set_ylim([0,2])
title('Señal con ruido, dominio temporal')

figure(6)
plot(fft.fftshift(k),abs(fft.fftshift(utn))/max(abs(fft.fftshift(utn))))
gca().set_xlim([-25,25])
gca().set_ylim([0,1])
xlabel('wavenumber (k)'); ylabel('|ut|/max(|ut|)')
title('Señal con ruido, dominio frecuencial')

#Filter:
k_0=1.5
tau=0.04
filter=exp(-tau*(k-k_0)**2)
unft=filter*utn
unf=fft.ifft(unft)

figure(7)
plot(k,utn)
title('SCISSORS')
plt.close()

figure(8)
plot(t,unf,'k')
gca().set_xlim([-25,25])
xlabel('Tiempo');ylabel('|u|')
title('Señal filtrada con filtro gaussiano con $\\tau=%0.2f$, $k_0=%0.2f$' % (tau,k_0))

figure(9)
plot(t,u)
gca().set_xlim([-25,25])
xlabel('Tiempo');ylabel('|u|')
title('Señal sin ruido blanco')
```

En este caso, se indujo un ruido conocido comunmente como "ruido blanco" en una señal de tipo sech(t). El filtro de esta señal se realizó usando un filtro gaussiano. Este filtro borra ciertas frecuencias aleatorias asumiendo una distribución normal desfasada, particularmente se define como:$$Fl(k)=e^{-\tau(k-k_0)^2}$$
Es posible observar que aunque la señal filtrada presenta algunas diferencias respecto a la señal real, se asemejan bastante, sobretodo en la parte en la cual la señal es concava hacia abajo, lo que, en teoría prueba la efectividad de este filtro.

#### Influencia del tiempo de muestreo
Dado que en la práctica, la transformación de Fourier no se aplica en todo el dominio de la función a evaluar,  es necesario tener en cuenta el tiempo de muestreo en el cual se aplicará la transformada. Normalmente, un tiempo de muestreo mayor implica una mejor reproducción de la señal

```python
#Radar detection and averaging
L=30; n=512; t2=linspace(-L,L,n); t=t2[:-1]
a=linspace(1,n/2,n/2-1); b=linspace(-n/2,-1,n/2); k=(2*pi/L)*concatenate((a,b),axis=0)
ks=fft.fftshift(k); noise=10
realize=[1,2,5,100]

for jj in realize:
    utn=np.zeros((jj,n-1))
    dat=np.zeros((jj,n-1))
    un=np.zeros((jj,n-1))
    u=1/cosh(t); ave=np.zeros(n-1)
    ut= fft.fft(u)
    for j in range (jj):
        utn[j,:]=ut+noise*(numpy.random.rand(n-1)+1j*numpy.random.rand(n-1))
        ave=ave+utn[j,:]
        dat[j,:]=abs(fft.fftshift(utn[j,:]))/max(abs(utn[j,:]))
        un[j,:]=fft.ifft(utn[j,:])
    ave=abs(fft.fftshift(ave))/jj
    figure()
    plot(ks,ave/max(ave))
    ylabel('|fft(u)|');xlabel('Frecuencia (k)')
    title('Número de datos ='+str(jj))
```
