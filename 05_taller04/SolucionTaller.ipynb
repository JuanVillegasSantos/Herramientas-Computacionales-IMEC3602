```python
L=20
n=128
x2=linspace(-L/2,L/2,n+1)
x=x2[:-1]
u=exp(-(x**2))
ut=fft.fft(u)
utshift=fft.fftshift(ut)

figure(1)
plot(x,u)
figure(2)
plot(ut)
figure(3)
plot(utshift)
plt.close()
```

```python

dx=x[1]-x[0]
u=1/cosh(x)
ut=fft.fft(u)
a=linspace(0,n/2-1,n/2)
b=linspace(-n/2,-1,n/2)

#FFT derivatives
k=(2*pi/L)*concatenate((a,b),axis=0)
ut1=k*ut*1j
ut2=-k*k*ut
u1=fft.ifft(ut1).real; u2=fft.ifft(ut2).real 
u1Exact=-tanh(x)/cosh(x); u2Exact= (1/cosh(x))-(2/(cosh(x))**3)

#Finite difference

#2nd order accuracy
ux=linspace(0,n,n); ux[0]=(-3*u[0]+4*u[1]-u[2])/(2*dx)
for i in range (1,n-2):
    ux[i]=(u[i+1]-u[i-1])/(2*dx)
ux[n-1]=(3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx)

#4th order accuracy
ux2=linspace(0,n,n)
ux2[0]=(-3*u[0]+4*u[1]-u[2])/(2*dx); ux2[1]=(-3*u[1]+4*u[2]-u[3])/(2*dx)
for i in range (2,n-3):
    ux2[i]=(-u[i+2]+8*u[i+1]-8*u[i-1]+u[i-2])/(12*dx)
ux2[n-2]=(3*u[n-2]-4*u[n-3]+u[n-4])/(2*dx)
ux2[n-1]=(3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx)

#Comparisson:

figure(1)
plot(x,u,'r', label='$f(x)=sech(x)$')

plot(x,u1,'g', label='$f´_{fft}(x)$=-tanh(x)sech(x)')
plot(x,u1Exact,'go',label='f´(x)=-tanh(x)sech(x)')

plot(x,u2,'b', label='$f´´_{fft}(x)$=sech(x)-$2sech^3(x)$')
plot(x,u2Exact,'bo',label='f´´(x)=sech(x)-$2sech^3(x)$')
legend()
```
