```python
import numpy as np
from scipy import integrate as inte
import pandas as pd
import sys
import matplotlib.pyplot as plt
```

Variables y ktes

```python
g=9.82 #gravity[m/s^2]
```

ISA (Atmósfera estandar)

```python
a=-6.5 #ISA gradient [K/km]
T0=288.15 #ISA Base temperature
R=287 # Ideal gas kte [m^2/(s^2-K)]
h0=0 #ISA Base heigh
rho=1.223 #ISA Base density
```

Masa equivalente

```python
mKart=152 #Masa del kart [kg]
mt=1.43
md=1.2
meq=mKart+mt+md

Nc=11/39 #Relación de transmisión
eficiencia=0.95 #Eficiencia de transmisión
rLl=0.139 #Radio geométrico llantas traseras [m]
cg=[-0.2231,0.4875,0.07478] #Centro de gravedad [-x,y,z] [m]
l=1.0063 #Batalla [m]
deltat=0.001 #[s]
```

Fuerzas

```python
def peso(m):
    w=m*g
    return w

def FInclinacion(m,theta):
    Fincl=peso(m)*np.sin(theta)
    return Fincl

def Drag(v,omega):
    d_del=0.975 #Distancia entre llantas delanteras [m]
    h=0.88 #Altura con el piloto sentado
    af=(d_del*h)/2 #Área frontal del kart
    cd=0.89 #kart Drag coeficient
    r_et=0.98*rLl
    vomega=omega*r_et
    D=(1/2)*rho*af*cd*np.math.fabs(v+vomega)**2
    return D

#Se modela la rodadura como mRod*a+kteRod
def Rodadura(m,v,theta):
    A=peso(m)*(cg[0]/l)
    B=peso(m)*(1/g*(cg[2]/l)) 
    fr=0.01*(1+(v/100))# Factor de rodadura velocidad en [km/h]
    C=fr*np.cos(theta) 
    mRod=B*C
    kteRod=A*C
    return [mRod,kteRod]

def FTorsion(Te):
    Ng=Nc
    ef=eficiencia
    r_et=0.98*rLl #Radio efectivo llantas traseras
    Ftor=Te*Ng*ef/r_et
    return Ftor

def Ffreno(coef,m,v,theta):
    A=peso(m)*(cg[0]/l)
    B=peso(m)*(1/g*(cg[2]/l)) 
    C=coef*np.cos(theta) 
    mFr=B*C
    kteFr=A*C
    return [mFr,kteFr]
```

Solución ecuación diferencial

Arranque:

```python
def dvdtA(v,t0):
    x,vel=v
    #Constantes e inputs funciones
    omega=5000*2*np.pi/60 #Velocidad angular
    m=meq #Masa del kart
    theta=0 #Ángulo de inclinación de la pista
    mRod=Rodadura(m,vel,theta)[0] #Masa del modelo de rodadura
    kRod=Rodadura(m,vel,theta)[1] #Constante del modelo de Rodadura
    #Torque en el eje
    Te=(3.037/omega)*((-0.00000004564538677*omega**3)+(0.0001056995459*omega**2)+(1.53209191*omega)-1.895523696)
    #Función a solucionar
    acel=(1/(m+4*mRod))*(FTorsion(Te)-4*kRod-Drag(vel,omega)-FInclinacion(m,theta))
    return [vel,acel]
def pos_vel(dvdt,frontera,t):
    sol=inte.odeint(dvdt,frontera,t)
    return sol
```

```python
# Variables de tiempo
t_inA=0
t_finA=10

# condiciones iniciales
x_0A=0
v_0A=0
fronteraA = [x_0A,v_0A]
tA=np.arange(t_inA,t_finA,deltat)
```

```python
solA=pos_vel(dvdtA,fronteraA,tA)
```

Frenado:

```python
def dvdtFr(v,t0):
    x,vel=v
    #Constantes e inputs funciones
    omega=5000*2*np.pi/60 #Velocidad angular
    m=mKart #Masa del kart
    theta=0 #Ángulo de inclinación de la pista
    FrCoef=0.85 #Friction coefficient
    mFr=Ffreno(FrCoef,m,vel,theta)[0] #Masa del modelo de rodadura
    kFr=Ffreno(FrCoef,m,vel,theta)[1] #Constante del modelo de Rodadura
    #Torque en el eje
    #Te=(3.037/omega)*((-0.00000004564538677*omega**3)+(0.0001056995459*omega**2)+(1.53209191*omega)-1.895523696)
    #Función a solucionar
    acel=(1/(m-mFr))*(-kFr+Drag(vel,omega)+FInclinacion(m,theta))
    return [vel,acel]
```

```python
# Variables de tiempo
t_inFr=0
t_finFr=10

# condiciones iniciales
x_0Fr=solA[-1,0]
v_0Fr=solA[-1,1]
fronteraFr = [x_0Fr,v_0Fr]
tFr=np.arange(t_inFr,t_finFr,deltat)
```

```python
solFr=pos_vel(dvdtFr,fronteraFr,tFr)
```

```python
pos=-1*np.append(solA[:,0],solFr[:,0])
velocity=-1*np.append(solA[:,1],solFr[:,1])
data=pd.DataFrame=({'Time':np.arange(0,20,deltat),'Position':pos,'Velocity':velocity})
pltPosition=data['Position'][data['Velocity']>=0]
pltVelocity=data['Velocity'][data['Velocity']>=0]
t=data['Time'][data['Velocity']>=0]

fig,ax1=plt.subplots()
ax2=ax1.twinx()

ax1.plot(t,pltPosition,'g-')
ax2.plot(t,pltVelocity,'b-')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Position [m]', color='g')
ax2.set_ylabel('Velocity [m/s]', color='b')
```
