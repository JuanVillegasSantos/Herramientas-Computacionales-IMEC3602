# Segmentación de Imágenes Usando la Transformada Watershed

La *Transformada Watershed* es un algortimo que permite realizar segmentación de imágenes. Así como el algoritmo Canny que trabajamos la clase pasada, es un paso preliminar a la aplicación de la transformada de Hough para la parametrización de los segmentos identificados.

# Imagen 1

```python
ret, thresh = cv2.threshold(Im1Gray,70,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, complement = cv2.threshold(Im1Gray,70,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
# noise removal
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
# sure background area
sure_bg = cv2.dilate(opening,kernel,iterations=3)
# Finding sure foreground area
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
```

```python
figure(3)
fig,axes= plt.subplots(nrows=2, ncols=2, figsize=(20, 16),
                                    sharex=True, sharey=True)
axes[0,0].imshow(Im1);axes[0,0].axis('off');axes[0,0].set_title('Original', fontsize=15)
axes[0,1].imshow(thresh,'gray');axes[0,1].axis('off');axes[0,1].set_title('binary image', fontsize=15)
axes[1,0].imshow(complement,'gray');axes[1,0].axis('off');axes[1,0].set_title('complement of image (a)', fontsize=15)
axes[1,1].imshow(dist_transform);axes[1,1].axis('off');axes[1,1].set_title('distance transform', fontsize=15)
```

```python
ret, sure_fg = cv2.threshold(dist_transform,0.3*dist_transform.max(),255,0)
# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)
# Add one to all labels so that sure background is not 0, but 1
markers = markers+1
# Now, mark the region of unknown with zero
markers[unknown==255] = 0
markers = cv2.watershed(Im1,markers)
```

```python
figure(3)
fig,axes= plt.subplots(nrows=5, ncols=2, figsize=(20, 18),
                                    sharex=True, sharey=True)
axes[0,0].imshow(Im1);axes[0,0].axis('off');axes[0,0].set_title('Original', fontsize=15)
Im1[markers == -1] = [255,0,0]
axes[0,1].imshow(thresh,'gray');axes[0,1].axis('off');axes[0,1].set_title('binary image', fontsize=15)
axes[1,0].imshow(complement,'gray');axes[1,0].axis('off');axes[1,0].set_title('complement of image (a)', fontsize=15)
axes[1,1].imshow(opening,'gray');axes[1,1].axis('off');axes[1,1].set_title('opening', fontsize=15)
axes[2,0].imshow(sure_bg);axes[2,0].axis('off');axes[2,0].set_title('Sure background', fontsize=15)
axes[2,1].imshow(dist_transform);axes[2,1].axis('off');axes[2,1].set_title('distance transform', fontsize=15)
axes[3,0].imshow(sure_fg);axes[3,0].axis('off');axes[3,0].set_title('sure foreground', fontsize=15)
axes[3,1].imshow(unknown);axes[3,1].axis('off');axes[3,1].set_title('unknown region', fontsize=15)
axes[4,0].imshow(markers);axes[4,0].axis('off');axes[4,0].set_title('markers', fontsize=15)
axes[4,1].imshow(Im1);axes[4,1].axis('off');axes[4,1].set_title('segmentation result', fontsize=15)
```
